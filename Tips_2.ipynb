```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

data = pd.read_csv("/content/tips.csv")

print(data.head())
```

```python
# Check data dimensions (rows & columns)
print(f"Dataset Shape: {data.shape}")

```

```python
# Check data types and missing values
print("\nData Info:")
print(data.info())
```

```python
# Check basic statistics for numerical columns
print("\nSummary Statistics:")
print(data.describe())
```

```python
# Scatter plot of total bill vs. tip
plt.figure(figsize=(5,3))
sns.scatterplot(x=data["total_bill"], y=data["tip"], hue=data["sex"])
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Total Bill vs Tip")
plt.show()
```

```python
# Compute correlation matrix
corr_matrix = data.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(6,3))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Box plot of tip amounts by day
plt.figure(figsize=(5,3))
sns.boxplot(x="day", y="tip", data=data)
plt.xlabel("Day of the Week")
plt.ylabel("Tip ($)")
plt.title("Tip Distribution by Day")
plt.show()
```

```python
# Histogram of tip amounts
plt.figure(figsize=(5,3))
sns.histplot(data["tip"], bins=10, kde=True)
plt.xlabel("Tip Amount ($)")
plt.ylabel("Frequency")
plt.title("Tip Amount Distribution")
plt.show()
```
