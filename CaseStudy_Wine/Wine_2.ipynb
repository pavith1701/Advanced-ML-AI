```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
data = pd.read_csv('/content/WineQT.csv')

print(data.head())
```

```python
# Check dataset dimensions
print(f"Dataset Shape: {data.shape}")

# Check column data types & missing values
print("\nData Info:")
print(data.info())

print("\nSummary Statistics:")
print(data.describe())

print("\nQuality Distribution:")  # wine quality counts
print(data["quality"].value_counts())
```

```python
# Plot bar chart for quality distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x="quality", hue="quality", palette="viridis", legend=False)
plt.xlabel("Wine Quality")
plt.ylabel("Count")
plt.title("Distribution of Wine Quality")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=data, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Boxplot of Wine Features")
plt.show()
```

```python
# Alcohol vs Quality

plt.figure(figsize=(8, 5))
sns.scatterplot(x=data["alcohol"], y=data["quality"], hue=data["quality"], palette="coolwarm")
plt.xlabel("Alcohol Content")
plt.ylabel("Wine Quality")
plt.title("Alcohol vs Wine Quality")
plt.show()
```
